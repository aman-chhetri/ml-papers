**1. ReLU (Rectified Linear Unit) :**

The Rectified Linear Unit (ReLU) or rectifier activation function introduces the property of nonlinearity to a deep learning model and solves the vanishing gradients issue.

It interprets the positive part of its argument. It is one of the most popular activation functions in deep learning.

The other variants of ReLU includes:

* `leaky ReLU`
* `p ReLU` (Parametric ReLU)
* `ELU` (Exponential Linear Unit)

which are used to improve performances in some tasks.


---




image.png

This function, `relu`, implements the ReLU activation function. The ReLU function outputs:

`ùë• if ùë• > 0`

`0 if ùë• ‚â§ 0`

---

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return np.maximum(x,0)

x=np.linspace(-10,10)
y=relu(x)
print(y)
plt.title("ReLU")
plt.plot(x,y)
```

**Note:**

* *The function is defined using `np.maximum(x, 0)`, which returns the element-wise maximum of the input `x` and `0`. For values of `x` less than or equal to zero, it returns zero; otherwise, it returns the value of `x`.*

* *`np.linspace(-10, 10)` generates an array of evenly spaced values of `x` from -10 to 10. By default, it generates 50 points.*

* *`y = relu(x)` applies the ReLU function to each element in the array `x`, producing an array `y` of ReLU values.*

**1.a) Leaky ReLU :**

It is an attempt to solve the dying ReLU problem. The range of the Leaky ReLU is (-infinity to infinity).

This `leaky`function returns:

`ùë• if ùë• > 0`

`0.1 * ùë• if ùë• ‚â§ 0`

image.png

```python
def leakyrelu(x):
  return np.maximum(x,0.1*x)

x=np.linspace(-10,10)
y=leakyrelu(x)
print(y)
plt.title("Leaky ReLU")
plt.plot(x,y)
```

**Note:**

* *The Leaky ReLU function introduces a small slope for negative values (here, 0.1), unlike the standard ReLU, which outputs zero for all negative inputs.*

* *This "leak" helps avoid the dying ReLU problem, where neurons can become inactive and stop learning during training if their gradient becomes zero. The small slope for negative values keeps the gradient alive even for negative inputs, allowing for some learning.*


---



**1.b) p ReLU (Parametric ReLU) :**

Parametric ReLU (PReLU) is an advanced variation of the traditional ReLU and Leaky ReLU activation functions, designed to further optimize neural network performance.

PReLU is an extension of the Leaky ReLU function, where the slope for negative values is not fixed but rather a parameter, ùõº, that can be learned during training.

The `p-ReLu` function returns:

`ùë• if ùë• > 0`

`ùõº √ó ùë• if  ùë• ‚â§ 0`

```python
def prelu(x,alpha):
  return np.maximum(x,alpha*x)

x=np.linspace(-10,10)
y=prelu(x,0.5)
print(y)
plt.title("PR ReLU (alpha=.5)")
plt.plot(x,y)
```

**Note:**

* *`alpha`: the slope parameter for the negative part of the function.*


* *`y = prelu(x, 0.5)` applies the PReLU function to each element of `x` with `alpha` set to 0.5. Here, for negative `x` values, the output is `0.5 * x`.*

**1.c) eLU (Exponential Linear Unit)** :

eLU is a function that tend to converge cost to zero faster and produce more accurate results. Different to other activation functions, eLU has a extra alpha constant which should be positive number.

The `eLU` funtion returns:

`ùë• if ùë• ‚â• 0`

`ùõº √ó (exp(ùë•)‚àí1) if  ùë• < 0`

image.png

```python
def elu(x,alpha):
  return np.where(x >= 0,x,alpha*(np.exp(x)-1))

x=np.linspace(-10,10)
y=elu(x,0.2)
print(y)
plt.title("eLU")
plt.plot(x,y)
```

**Note:**

* `y = elu(x, 0.2)` applies the ELU function with an `alpha` of 0.2, scaling the exponential curve for negative values.

* This is implemented using `np.where(ùë• >= 0 , ùë•, alpha*(np.exp(ùë•)-1))`, which evaluates each element of `ùë•` to check if it is positive or negative. For positive `ùë•`, it simply returns `ùë•`; for negative `ùë•`, it returns an exponentially decaying value scaled by  `alpha`.

**2. Softmax Function:**


The Softmax function is commonly used in the output layer of classification models because it converts a vector of values into probabilities. Each output value is in the range
[0,1], and the values sum up to 1.


image.png

```python
import numpy as np
import matplotlib.pyplot as plt

def softmax(x):
    exp_x = np.exp(x - np.max(x))  # Subtract max for numerical stability
    return exp_x / exp_x.sum(axis=0)

# Define a sample vector of values
x = np.linspace(-10, 10)  # A vector from -2 to 2
y = softmax(x)
print(y)

# Plotting Softmax
plt.title("Softmax")
plt.plot(x,y)
```

**Note:**

* *`exp_x = np.exp(x - np.max(x))` subtracts the maximum value of `x` to improve numerical stability and prevent overflow.*

* *`exp_x / exp_x.sum(axis=0)`normalizes the exponential values, resulting in a probability distribution.*

**3. Tanh Function**

The Tanh (Hyperbolic Tangent) function is another commonly used activation function, especially in hidden layers of neural networks. It outputs values between -1 and 1.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)

# Define x range
x = np.linspace(-10, 10)
y = tanh(x)
print(y)
# Plotting Tanh
plt.title("Tanh ")
plt.plot(x,y)
```

**4. Sigmoid Function:**

The Sigmoid activation function is another widely used function, especially in binary classification problems. It maps any input to a range between 0 and 1, making it useful for modeling probabilities.

image.png

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Define x range
x = np.linspace(-10, 10)
y = sigmoid(x)
print(y)

# Plotting Sigmoid
plt.title("Sigmoid")
plt.plot(x,y)
```

**Note:**

* *`1 / (1 + np.exp(-x))` applies the Sigmoid function to each element in `x`.*

* *Plot shows the characteristic "S-shaped" curve of the Sigmoid function, smoothly transitioning from 0 to 1 as `x` moves from negative to positive values.*

**Extra Tips for you:**

**Choosing the Right Activation Function**: The choice of activation function depends on the type of problem you are trying to solve.

Here are some guidelines:

* **For binary classification:** Use the  `sigmoid` activation function in the output layer. It will squash outputs between 0 and 1, representing probabilities for the two classes.

*   **For multi-class classification:** Use the `softmax` activation function in the output layer. It will output probability distributions over all classes.

* **If unsure**: Use the `ReLU` activation function in the hidden layers. ReLU is the most common default activation function and usually a good choice.
