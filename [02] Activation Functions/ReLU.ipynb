```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
  return np.maximum(x,0)

x=np.linspace(-10,10)
y=relu(x)
print(y)
plt.title("ReLU")
plt.plot(x,y)
```

```python
def leakyrelu(x):
  return np.maximum(x,0.1*x)

x=np.linspace(-10,10)
y=leakyrelu(x)
print(y)
plt.title("Leaky ReLU")
plt.plot(x,y)
```

```python
def prelu(x,alpha):
  return np.maximum(x,alpha*x)

x=np.linspace(-10,10)
y=prelu(x,0.5)
print(y)
plt.title("PR ReLU (alpha=.5)")
plt.plot(x,y)
```

```python
def elu(x,alpha):
  return np.where(x >= 0,x,alpha*(np.exp(x)-1))

x=np.linspace(-10,10)
y=elu(x,0.2)
print(y)
plt.title("eLU")
plt.plot(x,y)
```
